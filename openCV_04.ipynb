```python
import cv2
import numpy as np
import sys
```

```python
src = cv2.imread('fig/puppy.bmp')
sunglass = cv2.imread('fig/imgbin_sunglasses_1.png',cv2.IMREAD_UNCHANGED) # IMREAD_UNCHANGED가 없으면 alpha channel 입력 안됨
 
mask = sunglass[:,:,-1] # 이미지의 alpha channel을 mask로 사용
glass = sunglass[:,:,0:3] # RGB까지만 입력

mask = cv2.resize(mask,(300,150)) # mask파일의 크기 조절
glass = cv2.resize(glass,(300,150)) # src 파일의 크기 조절

h,w = mask.shape[:2]
src_crop = src[120:120+h, 220:220+w] # src의 원본데이터 수정

cv2.copyTo(glass, mask, src_crop)    

cv2.imshow('src',src)
cv2.imshow('mask',mask)
cv2.imshow('glass',glass)
cv2.imshow('src_crop',src_crop)

cv2.waitKey()
cv2.destroyAllWindows()
```

### 이미지의 alpha channel을 mask로 바꿔 이미지 변경

```python
src = cv2.imread('fig/cat.bmp')
bin_hat = cv2.imread('fig/imgbin_hat.png',cv2.IMREAD_UNCHANGED)
 
mask = bin_hat[:,:,-1]
hat = bin_hat[:,:,0:3]

mask = cv2.resize(mask,(300,400))
hat = cv2.resize(hat,(300,400))

h,w = mask.shape[:2]

src_crop = src[0:0+h, 200:200+w]

cv2.copyTo(hat, mask, src_crop) 

cv2.imshow('src',src)
cv2.imshow('mask',mask)
cv2.imshow('hat',hat)
cv2.imshow('src_crop',src_crop)

cv2.waitKey()
cv2.destroyAllWindows()
```

### 여러가지 선 및 도형 그려보기

```python
img = np.full((800,1200,3), (255,255,255), dtype=np.uint8)

#line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
# @param img Image.
# .   @param pt1 First point of the line segment.
# .   @param pt2 Second point of the line segment.
# .   @param color Line color.
# .   @param thickness Line thickness.
# .   @param lineType Type of the line. See #LineTypes.
# .   @param shift Number of fractional bits in the point coordinates.

cv2.line(img,(50,50),(300,50), (0,0,255), 10)
cv2.line(img,(300,50),(200,300), (0,0,255), 10)
cv2.line(img,(400,50),(400,300), (255,0,0), 10)

# rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img # pt1 : 좌측 상단, pt2 : 우측 하단
# rectangle(img, rec, color[, thickness[, lineType[, shift]]]) -> img

cv2.rectangle(img, (50, 400),(400,600),(0,0,255),-1)
cv2.rectangle(img, (60,450,250,100),(255,255,255),-1) # rec : 사각형 위치 정보 (x, y, w, h)

# circle(img, center, radius, color[, thickness[, lineType[, shift]]]) -> img
cv2.circle(img, (600,300), 100, (0,0,255), 10, cv2.LINE_AA) # cv2.LINE_AA : 굴곡을 그릴때 부드럽게 그림

# ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img
cv2.ellipse(img, (800, 300), (50,100), 30, 0, 340,(127,0,255), 10, cv2.LINE_AA) # center : 원의 중심좌표, axes :축의 반지름

# putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]) -> img
cv2.putText(img,'OpenCV', (600, 500), cv2.FONT_HERSHEY_COMPLEX, 2,
        (0,0,0), 5, cv2.LINE_AA)


cv2.imshow('img',img)

cv2.waitKey()
cv2.destroyAllWindows()
```

### circle, line, rectangle을 이용하여 글자 써보기 (인삼)

```python
img = np.full((500,500,3),(255,255,255),dtype=np.uint8)

# 인
cv2.circle(img, (170,200), 30, (0,0,255), 5, cv2.LINE_AA)
cv2.line(img, (220,150), (220,255), (0,255,0),5)
cv2.line(img, (150,250), (150,275), (0,255,0),5)
cv2.line(img, (150,275), (220,275), (0,255,0),5)

# 삼
cv2.line(img, (290,170), (260,220), (0,255,0),5)
cv2.line(img, (290,170), (320,220), (0,255,0),5)

cv2.line(img, (340,160), (340,240), (0,255,0),5)
cv2.line(img, (340,200), (370,200), (0,255,0),5)

cv2.rectangle(img, (280,250,60,30),(0,255,255),5) # rec : 사각형 위치 정보 (x, y, w, h)

cv2.imshow('img',img)

cv2.waitKey()
cv2.destroyAllWindows()
```

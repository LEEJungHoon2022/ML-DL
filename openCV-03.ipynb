```python
import cv2
import numpy as np
import sys
```

```python
img1 = cv2.imread('fig/puppy.bmp')

if img1 is None:
    print('image read failed')
    sys.exit()

img2 = img1[200:400, 300:500] # 주소 복사
img3 = img1[200:400, 300:500].copy() # 원본 데이터 복사

# circle(img, center, radius, color[, thickness[, lineType[, shift]]]) -> img
cv2.circle(img2,(100,100), 50, (255,255,0), 1, cv2.LINE_AA) # img2 100,100에 크기가 50인 원의 색을 (255,255,0)으로 지정 후 원 두께를 1로 지정하여 그림

cv2.imshow('img2',img2) # 주소값을 불러와서 원본데이터도 수정됨
cv2.imshow('img1',img1) # 주소값으로 부른 변수를 수정하니 해당 영역이 같이 수정 됨
cv2.imshow('img3',img3) # 데이터를 저장해서 원본 데이터와 무관하여 수정 안됨

cv2.waitKey()
cv2.destroyAllWindows()
```

### mask 파일을 만들어 다른 영상 입력하기

```python
src = cv2.imread('fig/cow.png')
dst = cv2.imread('fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY) # mask 파일을 생성하기 위해 회색으로 src파일을 입력

if src is None or dst is None:
    print('image read failed')
    sys.exit()

#threshold(src, thresh, maxval, type[, dst]) -> retval, dst // thresh , maxval : thresh ~ maxval 사이의 값이 아니면 type으로 색을 바꿔라
ret, src_mask = cv2.threshold(src_gray, 240,255, cv2.THRESH_BINARY_INV) # src_gray의 영상의 240 ~ 255 사의의 값이 아닌 부분을 검은색(THRESH_BINARY)으로 바꿔라 그런데 색을 INV(반전)해라


h, w = src.shape[:2] # src와 dst의 크기가 다르기 때문에 copyto를 할 수 없어 크기를 조절
# print(h, w) 크기 확인

dst_resize = cv2.resize(dst,(w,h), interpolation=cv2.INTER_AREA)

# copyTo(src, mask[, dst]) -> dst // src파일을 mask하여 dst파일에 저장함
cv2.copyTo(src,src_mask, dst_resize)
    
cv2.imshow('src',src)
# cv2.imshow('src_gray',src_gray)
cv2.imshow('src_mask',src_mask)
# cv2.imshow('dst',dst)
cv2.imshow('dst_resize',dst_resize)

cv2.waitKey()
cv2.destroyAllWindows()
```

### dst파일의 크기를 조절하지 않고 mask하기

```python
src = cv2.imread('fig/cow.png')
dst = cv2.imread('fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)


# print(src.shape[:2]) # src의 크기 확인(h = 600 ,w = 600)
h, w = src.shape[:2] # src의 h, w를 입력
# print(dst.shape[:2]) # dst의 크기 확인(h = 1125, w = 1500)
n_dst = dst[525:525+h ,400:400+w] # 원본 데이터 주소를 불러와 위치 지정

if src is None or dst is None or src_gray is None:
    print('image read failed')
    sys.exit()
ret, src_mask = cv2.threshold(src_gray, 240,255, cv2.THRESH_BINARY_INV)

cv2.copyTo(src,src_mask, n_dst) # n_dst를 mask파일로 수정하여 src영상을 입력하면 원본데이터의 주소를 가져왔기 때문에 원본 데이터도 같이 수정됨 
    
cv2.imshow('dst',dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

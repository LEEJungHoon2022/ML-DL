```python
img = cv2.imread('fig/puppy.bmp',cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()

cv2.namedWindow('image',cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',img)

while True:
    if cv2.waitKey() == 27: # ASCII 27 == ESC
        break
        
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
imgGRAY = cv2.imread('fig/puppy.bmp',cv2.IMREAD_GRAYSCALE) # cv2는 BGR 순서로 파일을 일기때문에 matplot은 색깔이 다르게 출력

if img is None:
    print('image read failed')
    sys.exit()
    
# imgBGR = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB) # COLOR_BGR2RGB : BGR순인 데이터를 RGB순으로 읽게해줌
    
plt.imshow(imgGRAY, cmap='gray') # sudo gray // cmap = 'gray' 로 해야 정상적으로 색깔 출력
plt.axis('off')
plt.show()
```

```python
imgBGR = cv2.imread('fig/puppy.bmp', cv2.IMREAD_COLOR)
imgGRAY = cv2.imread('fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

if imgBGR is None or imgGRAY is None:
    print('image read falied')
    sys.exit()

imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(20,20))
plt.subplot(131), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(132), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(133), plt.imshow(imgGRAY, cmap='gray'), plt.axis('off')
plt.show()
```

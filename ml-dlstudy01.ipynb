# 히스토그램(histplot)

```python
sns.histplot(data = titanic, x = 'age', bins = 10) # bins : 구간 개수 설정 default : auto
```

```python
sns.histplot(data = titanic, x = 'age', hue = 'alive' ) 
# hue : (나이)빈도를 특정 범주(alive)별로 구분하여 볼 때 사용, # 포개져서 그려지기 때문에 회색 부분이 겹친 부분
```

```python
sns.histplot(data = titanic, x = 'age', hue = 'alive', multiple = 'stack') 
# multiple = stack : hue로 그래프를 그릴때 포개지 않고 그림(누적해서 표현)
```

```python
sns.kdeplot(data = titanic , x = 'age') 
# 커널 밀도 측정 쉽게 생각해 히스토그램을 매끄러운 곡선으로 연결한 그래프 정도로 이해 / 실제로 탐색적 데이터 분석 시 많이 쓰이지는 않음
# 앞서 그린 히스토그램과 비교해보면 히스토그램을 매끈하게 연결한 모습인것을 확인 가능 / 히스토그램은 이산적이지만 커널밀도추정 함수 그래프는 연속적
```

```python
sns.kdeplot(data = titanic, x = 'age', hue = 'alive', multiple = 'stack')
# hue 와 multiple parameter 사용 데이터가 연속적이다는 특징 제외하고는 모양새는 비슷함
```

# 분포도(displot)
#### 수치형 데이터 하나의 분포를 나타내는 그래프
#### 캐글에서 분포도를 그릴때는 displot()을 많이 사용
#### parameter 만 조정하면 histplot 과 kdeplot이 제공하는 기본 그래프를 모두 그릴 수 있음

```python
sns.displot(data = titanic, x = 'age')
# parameter를 기본값으로 두면 히스토그램을 그림
# 앞에 히스토그램과 비교해보면 크기만 다름
```

```python
sns.displot(data = titanic, x = 'age', kind = 'kde')
# kind parameter를 'kde'로 설정하면 커널밀도추정 함수를 그림 / 위의 그래프와 크기만 다름
```

```python
sns.displot(data = titanic, x = 'age', kde = True)
# kde parameter를 True로 설정하면 히스토그램과 커널밀도추정 그래프 두개를 그릴 수 있음
```

# 러그플롯(rugplot)
#### 러그플롯은 주변 분포를 나타내는 그래프이다.
#### 단독으로 사용하기보다는 주로 다른 분포도 그래프와 함께 사용함

```python
sns.kdeplot(data = titanic, x = 'age')
sns.rugplot(data = titanic, x = 'age')
# 커널밀도추정 함수 그래프와 러그플롯을 함께 그린 예시
# 러그 플롯은 단일 피처(여기서는 age)가 어떻게 분포돼 있는지를 x축 위에 작은 선분(러그)으로 표시함
# 값이 밀집돼 있을수록 작은 선분들도 밀집돼 있음
```

# 범주형 데이터 시각화

## 막대 그래프(barplot)
#### 막대그래프는 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 그려줌
#### 수치형 데이터 평균은 막대 높이로, 신뢰구간은 오차 막대로 표현함
#### barplot은 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 그려줌
#### 수치형 데이터는 평균을 막대 높이로 신뢰구간은 오차 막대 (error bar)로 표현
#### 원본 데이터를 복원 샘플링하여 얻은 표본을 활용해 평균과 신뢰구간을 구하는 것임
#### 즉 barplot은 원본데이터 평균이 아니라 샘플링한 데이터의 평균을 구해줌
#### 기본적으로 x 파라미터에 범주형 데이터를 y 파라미터에 수치형 데이터를 전달함

```python
sns.barplot(x = 'class', y = 'fare', data = titanic)
# 막대 높이는 등급별 평균 운임을 뜻함
# 막대 상단의 검은색 세로줄이 오차막대(신뢰구간)이다.
# 등급이 높을수록 평균 운임이 비싸고 신뢰구간이 넓어짐을 확인가능함
```

# 포인트플롯(pointplot)
#### 막대그래프와 모양만 다를 뿐 동일한 정보를 제공함
#### 막대그래프와 마찬가지로 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 나타냄
#### 그래프를 점과 선으로 나타냄

```python
sns.pointplot(x='class', y='fare', data=titanic)
# 그래프를 보면 포인트플롯은 막대그래프와 동일한 정보 제공
# 한 화면에 여러그래프를 그릴때 사용
# 포인트 플롯은 점과 선으로 표현하기 때문에 여러 그래프를 그려도 서로 잘보이고 비교하기도 쉬움
```

# 박스플롯(boxplot)
#### 박스플롯은 막대 그래프나 포인트 플롯보다 더 많은 정보를, 구체적으로 5가지 요약 수치를 제공
#### 5가지 요약 수치는 최솟값, 제1사분위 수(Q1), 제2사분위수(Q2), 제3사분위 수(Q3), 최댓값을 뜻함

##### 제1사분위 수(Q1) 전체 데이터 중 하위 25%에 해당하는 값
##### 제2사분위 수(Q2) 50%에 해당하는 값(중앙값)
##### 제3사분위 수(Q3) 상위 25에 해당하는 값
##### 사분위 범위 수 (IQR) Q3 - Q1
##### 최댓값 Q3 + (1.5 * IQR)
##### 최솟값 Q1 - (1.5 * IQR)
##### 이상치 최댓값보다 큰 값과 최솟값보다 작은 값 

### boxplot의 x,y parameter에는 각각 범주형 데이터(class) 와 수치형 데이터(age)를 전달

```python
sns.boxplot(x = 'class', y = 'age', data = titanic)
```

# 바이올린플롯(violinplot)
### 박스플롯과 커널밀도추정 함수 그래프를 합쳐놓은 형태

```python
sns.violinplot(x = 'class', y = 'age', data = titanic)
# boxplot은 각 범주별 5가지 요약 수치를 한눈에 보고싶으면 사용
# violinplot 수치형 데이터의 전체적인 분포 양상을 알고싶다면 violinplot 사용
```

```python
sns.violinplot(x = 'class', y = 'age', hue = 'sex', data=titanic, split=True) # hue : 성별에 따른 나이 분포도 split=True : hue에 전달한 Feature를 반으로 나누어 보여줌
```

# countplot
### countplot은 범주형 데이터의 개수를 확인 할 때 사용하는 그래프 주로 범주형 Feature나 범주형 target값의 분포가 어떤지 파악하는 용도로 사용

```python
sns.countplot(x = 'class',data=titanic) # x parameter를 y로 변경하면 그래프의 방향을 바꿀수 있음
# 범주형 데이터 개수 파악 가능
```

```python
sns.countplot(y = 'class',data=titanic) # parameter y 
# 범주형 데이터 개수가 많아 그래프가 옆으로 너무 넓어져 보기 불편할 때 매우 유용
```

# 데이터 관계 시각화
### 관계도는 여러 데이터 사이의 관계를 살펴보기 위한 그래프이다.
### 히트맵, 라인플롯, 산점도, 회귀선을 포함한 산점도

### heatmap
#### heatmap은 데이터 간 관계를 색상으로 표현한 그래프임, 비교해야하는 데이터가 많을 때 주로 사용

```python
import seaborn as sns
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot(index='month',
                             columns='year',
                             values='passengers')
flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
# 전체적인 양상을 쉽게 볼 수 있음
# 표를 볼때보다 의미를 파악하기 쉬움
```

# lineplot
### line plot은 두 수치형 데이터 사이의 관계를 나타낼 때 사용
### 기본적으로는 x parameter에 전달한 값에 따라 y parameter에 전달한 값의 평균과 95% 신뢰구간을 나타냄

```python
sns.lineplot(x='year',y='passengers',data=flights)
# x축은 연도 y축은 평균 승객 수를 나타냄
# 실선 주변 음영은 95% 신뢰 구간
```

# scatter
### scatter는 두 데이터의 관계를 점으로 표현하는 그래프이다.
### scatter 그래프에는 총 비용과 팁 정보를 모아둔 tips 데이터셋을 활용

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
# 대채로 총액이 늘면 팁도 따라서 늘고 있음
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='time',data=tips)
# hue : 산점도를 특정 범주형 데이터별로 나누어 그릴 수 있음(점심과 저녁으로 나눠짐) 전체적인 추이는 비슷
```

# 회귀선을 포함한 산점도 그래프(regplot)
### regplot은 산점도와 선형 회귀선을 동시에 그려주는 함수
### 회귀선을 그리면 전반적인 상관관계 파악이 좀더 쉬움

```python
sns.regplot(x='total_bill',y='tip',data=tips) # 선형 회귀선 주변 음영은 95% 신뢰구간을 의미
```

```python
sns.regplot(x='total_bill',y='tip', ci=99, data=tips) # ci : 신뢰구간을 늘림 여기서는 99%로 늘렸음 음영이 늘어난것이 확인 가능
```

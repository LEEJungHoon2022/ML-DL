```python
import cv2
import numpy as np
import sys
import os
```

```python
img1 = cv2.imread('fig/puppy.bmp',cv2.IMREAD_COLOR)
img2 = cv2.resize(img1,(700,500)) # img1 을 resize

cv2.imwrite('fig/resize_p.bmp',img2) # resize한 데이터를 저장

cv2.imshow('img1',img1)
cv2.imshow('img2',img2)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_list_os = os.listdir('fig/images/') # 이미지 데이터가 있는 폴더로 경로 설정 후 파일명을 리스트 형식으로 저장

files_list = []

for image in img_list_os:
    img_path = 'fig/images/' + image # 해당 이미지 경로 생성
    files_list.append(img_path)

# print(files_list)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL) # WINDOW_AUTOSIZE로 창을 생성하면 창 크기 조절 불가로 WND_PROP_FULLSCREEN, WINDOW_FULLSCREEN 명령어가 작동하지 않음
cv2.setWindowProperty('scene',cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN) # window 창 생성시 Fullscreen으로 생성 됨

idx = 0

while True:
    img = cv2.imread(files_list[idx], cv2.IMREAD_COLOR) # 파일을 0번 index부터 불러옴 ,색상은 COLOR로
    
    if img is None:
        print('image read failed')
        sys.exit()
    
    cv2.imshow('scene',img) # scene으로 생성한 Window 창에 출력함
    
    if cv2.waitKey(1500) == 27:
        break
    
    idx += 1
    
    if idx == len(files_list):
        idx = 0

cv2.destroyAllWindows()
```

```python
# zeros(shape, dtype=float, order='C', *, like=None)
img1 = np.zeros((400,500),dtype=np.uint8)
img2 = np.ones((400, 500)) * 255 # 인수를 3개로 넣어주면 3차원 RGB가 생성됨 2차원으로 생성하면 RGB가 없음 // dtype을 지정해주지 않으면 float 형식으로 생성 되기 때문에 np.uint8을 지정해줘야됨
img3 = np.full((500,600,3),(255,255,255), dtype=np.uint8) # dtype이 없으면 error 발생
img4 = np.random.randint(0,255, (500,600), dtype=np.uint8) # 영상을 랜덤값으로 넣어줘서 노이즈 영상 생성


cv2.imshow('img1',img1)
cv2.imshow('img2',img2)
cv2.imshow('img3',img3)
cv2.imshow('img4',img4)

cv2.waitKey()
cv2.destroyAllWindows()
```

## 문제 설명
#### 수많은 마라톤 선수들이 마라톤에 참여하였습니다. 단 한 명의 선수를 제외하고는 모든 선수가 마라톤을 완주하였습니다.
#### 마라톤에 참여한 선수들의 이름이 담긴 배열 participant와 완주한 선수들의 이름이 담긴 배열 completion이 주어질 때, 완주하지 못한 선수의 이름을 return 하도록 solution 함수를 작성해주세요.

## 제한사항
#### 마라톤 경기에 참여한 선수의 수는 1명 이상 100,000명 이하입니다.
#### completion의 길이는 participant의 길이보다 1 작습니다.
#### 참가자의 이름은 1개 이상 20개 이하의 알파벳 소문자로 이루어져 있습니다.
#### 참가자 중에는 동명이인이 있을 수 있습니다.
#### 입출력 예

participant	completion	returnb
["leo", "kiki", "eden"]	                             ["eden", "kiki"]                        : "leo"

["marina", "josipa", "nikola", "vinko", "filipa"]	["josipa", "filipa", "marina", "nikola"] : "vinko"

["mislav", "stanko", "mislav", "ana"]	            ["stanko", "ana", "mislav"]              : "mislav"

```python
# 해쉬를 이용한 풀이

def solution(participant,completion):
    
    dicts = {} # dictonary 생성
    sum_01 = 0
    sum_02 = 0
    
    for person in participant: # participant 리스트의 데이터가 하나씩 person으로 입략됨
        dicts[hash(person)] = person # 고유한 값을 생성해네는 메서드인 hash() 로 key 생성 후 value에는 이름(str)을 입력
        sum_01 += hash(person) # 모든 hash 값을 sum_01에 저장
    
    for person in completion: # completion의 데이터를 하나씩 person에 저장
        sum_02 += hash(person) # persom에 있는 데이터를 hash() 메서드를 이용하여 고유한 값으로 변경
    
    key = sum_01 - sum_02 # particpant의 데이터가 completion의 데이터보다 1개가 많은점을 이용하여 고유한 값을 모두 더한값에서 하나 적은 값으로 빼면 고유한 데이터가 하나 남음
                          
    
    answer = dicts[key] # 그깂이 key가 되어 value를 호출험
    
    return answer
a = ["leo", "kiki", "eden"]
b = ["eden", "kiki"]

solution(a,b) # leo가 return 됨
```
